## AULA 3:

### EXERCÍCIO: ANSCOMBE

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Anscombe_quartet_data.csv')
df
```

```python
x123 = df['x123']
x4 = df['x4']
y4 = df['y4']
y3 = df['y3']
y2 = df['y2']
y1 = df['y1']
```

```python
rgr_1 = LinearRegression()
x_1 = pd.DataFrame(df, columns=['x123'])
y_1 = pd.DataFrame(df, columns=['y1'])
rgr_1.fit(x_1, y_1)

rgr_2 = LinearRegression()
y_2 = pd.DataFrame(df, columns=['y2'])
rgr_2.fit(x_1, y_2)

rgr_3 = LinearRegression()
y_3 = pd.DataFrame(df, columns=['y3'])
rgr_3.fit(x_1, y_3)

rgr_4 = LinearRegression()
x_4 = pd.DataFrame(df, columns=['x4'])
y_4 = pd.DataFrame(df, columns=['y4'])
rgr_4.fit(x_4, y_4)
```

```python
color_1 = 'orange'
color_2 = 'blue'
sns.set_style('whitegrid')
plt.figure(figsize=[20, 10])

plt.subplot(2, 2, 1)
plt.title(f'Correlação = {round(x123.corr(y1), 3)}', fontsize=14)
plt.scatter(x123, y1, color=color_1, s=200)
plt.plot(x123, rgr_1.predict(x_1), color=color_2)

plt.subplot(2, 2, 2)
plt.title(f'Correlação = {round(x123.corr(y2), 3)}', fontsize=14)
plt.scatter(x123, y2, color=color_1, s=200)
plt.plot(x123, rgr_2.predict(x_1), color=color_2)

plt.subplot(2, 2, 3)
plt.title(f'Correlação = {round(x123.corr(y3), 3)}', fontsize=14)
plt.scatter(x123, y3, color=color_1, s=200)
plt.plot(x123, rgr_3.predict(x_1), color=color_2)

plt.subplot(2, 2, 4)
plt.title(f'Correlação = {round(x4.corr(y4), 3)}', fontsize=14)
plt.scatter(x4, y4, color=color_1, s=200)
plt.plot(x4, rgr_4.predict(x_4), color=color_2)
```

### EXERCÍCIO: TEMPORAL

```python
df = px.data.iris()
df
```

```python
fig = px.scatter_matrix(df, dimensions=["sepal_width", "sepal_length", "petal_width", "petal_length"], color="species")
fig.show()
```

```python
fig = px.parallel_coordinates(df, color='species_id', dimensions=["sepal_width", "sepal_length", "petal_width", "petal_length"])
fig.show()
```

```python
df
```

```python
features = df.loc[:,:'petal_width']
features
```

```python
tsne = TSNE(n_components=2)

projections = tsne.fit_transform(features)
projections

fig = px.scatter(projections, x=0, y=1, color=df.species)
fig.show()
```

```python
tsne = TSNE(n_components=3)

projections = tsne.fit_transform(features)
projections

fig = px.scatter_3d(projections, x=0, y=1, z=2, color=df.species)
fig.show()
```

```python
from dash import Dash, html, dcc, Input, Output
```

```python
app = Dash(__name__)

app.layout = html.Div([
    html.H4("PROGRESSÃO DA EXPECTATIVA DE VIDA DOS PAÍSES POR CONTINENTES"),
    dcc.Graph(id="graph"),
    dcc.Checklist(
        id="checklist",
        options=["Asia", "Europe", "Africa", "Americas", "Oceania"],
        value=["Europe", "Africa"],
        inline=True,
    ),
])

@app.callback(
    Output("graph", "figure"),
    Input("checklist", "value")
)

def update_line_chart(continents):
    df = px.data.gapminder()
    mask = df.continent.isin(continents)
    fig = px.line(df[mask], x="year", y="lifeExp", color="country", title="PROGRESSÃO DA EXPECTATIVA DE VIDA DOS PAÍSES POR CONTINENTES")
    return fig

app.run()
```
